```python
from urllib import request
url = "http://users.isr.ist.utl.pt/~cfb/VdS/v301.txt" #Amor é fogo que arde sem se ver
response = request.urlopen(url)
raw1 = response.read().decode('latin1')

url = "http://users.isr.ist.utl.pt/~cfb/VdS/v303.txt" #Verdes são os campos
response = request.urlopen(url)
raw2 = response.read().decode('latin1')

url = "http://users.isr.ist.utl.pt/~cfb/VdS/v305.txt" #Busque Amor novas artes, novo engenho

response = request.urlopen(url)
raw3 = response.read().decode('latin1')
tokens1=word_tokenize(raw1)
tokens2=word_tokenize(raw2)
tokens3=word_tokenize(raw3)

duplas_obra_tokens=[("Amor é fogo que arde sem se ver",w) for w in tokens1]+[("Verdes são os campos",w) for w in tokens2]+\
[("Busque Amor novas artes, novo engenho",w) for w in tokens3]
```

```python
duplas_obra_tokens_a=[(p,w.lower()) for (p,w) in duplas_obra_tokens]
```
