## Exercício 1
Considere o poema Trem de Ferro de Manuel Bandeira:

```python
poema="""Café com pão
Café com pão
Café com pão
Virge Maria que foi isto maquinista?

Agora sim
Café com pão
Agora sim
Voa, fumaça
Corre, cerca
Ai seu foguista
Bota fogo
Na fornalha
Que eu preciso
Muita força
Muita força
Muita força

Oô...
Foge, bicho
Foge, povo
Passa ponte
Passa poste
Passa pasto
Passa boi
Passa boiada
Passa galho
De ingazeira
Debruçada
No riacho
Que vontade
De cantar!

Oô...
Quando me prendero
No canaviá
Cada pé de cana
Era um oficiá

Oô...
Menina bonita
Do vestido verde
Me dá tua boca
Pra matá minha sede
Oô...
Vou mimbora vou mimbora
Não gosto daqui
Nasci no Sertão
Sou de Ouricuri
Oô...

Vou depressa
Vou correndo
Vou na toda
Que só levo
Pouca gente
Pouca gente
Pouca gente..."""
```

a) Realize a tokenização do poema

```python
from nltk import word_tokenize
tokens = word_tokenize(poema)
types = set(tokens)
tokens, types
```

b) Obtenha o número de tokens, types e a diversidade lexical

```python
num_tokens = len(poema)
num_tokens
```

```python
num_types = len(set(poema))
num_types
```

```python
div_lexical = num_types / num_tokens
div_lexical
```

c) Obtenha a distribuição da frequência das palavras do texto e faça um histograma. Dica: utilize o código abaixo.

```python
import nltk

fdist=nltk.FreqDist(tokens)
fdist_sort=fdist.most_common(num_types)
labels=[w for (w,f) in fdist_sort ] # cria uma lista dos types
freq=[f for (w,f) in fdist_sort ] # cria uma lista com as frequências
fdist.plot()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_pos = np.arange(len(freq))
f=plt.figure()
f.set_figwidth(30)
f.set_figheight(6)
# Create bars
plt.bar(x_pos, freq)
 
# Rotation of the bar names
plt.xticks(x_pos, labels, rotation=90)

plt.ylabel('Frequência')
plt.xlabel('Type')
# Show graphic
plt.show()
```

d) A partir do resultado do item c), obtenha quantas vezes as palavras "pão" e "passa" aparecem. Utilize para isso comandos do tipo fdist["pão"]


```python
num_pao = fdist["pão"]
num_passa = fdist["passa"]
num_Passa = fdist["Passa"]
num_pao, num_passa, num_Passa
```

e) Obtenha a probabilidade de um token do poema escolhido ao acaso ser a palavra "força".

```python
num_forca = fdist["força"]
prob = round((num_forca  * 100 / num_tokens), 2)
str(prob) + " %"
```

f) Obtenha uma lista de palavras do poema que terminam com o sufixo 'ista'

```python
lista_1 = []
for i in types:
    if i.endswith("ista"):
        lista_1.append(i)
lista_1
```

g) Obtenha uma lista de palavras do poema que iniciam com letra maiúscula.

```python
lista_2 = []
for i in types:
    if i[0].isupper():
        lista_2.append(i)
lista_2
```

h) Obtenha uma lista de palavras que possuam menos de 3 caracteres

```python
lista_3 = []
for i in types:
    if len(i) < 3:
        lista_3.append(i)
lista_3
```

```python
materia_tokens = word_tokenize(materia)
materia_types = set(materia_tokens)
materia_num_tokens = len(materia_tokens)
materia_num_types = len(materia_types)
div_lexical_materia = materia_num_types / materia_num_tokens
comp = div_lexical_materia / div_lexical # A diversidade lexical da matéria é cerca de 8 vezes maior que do Poema Trem de Ferro.

materia_num_tokens, materia_num_types, div_lexical_materia, comp
```

c) Obtenha a distribuição da frequência das palavras do texto e faça um histograma.

```python
fdist=nltk.FreqDist(materia_tokens)
fdist_sort=fdist.most_common(materia_num_types)
labels_materia=[w for (w,f) in fdist_sort ] # cria uma lista dos types
freq_materia=[f for (w,f) in fdist_sort ]

x_pos_materia = np.arange(len(freq_materia))
m=plt.figure()
m.set_figwidth(30)
m.set_figheight(6)
# Create bars
plt.bar(x_pos_materia, freq_materia)
 
# Rotation of the bar names
plt.xticks(x_pos_materia, labels_materia, rotation=90)

plt.ylabel('Frequência Materia')
plt.xlabel('Type Materia')
# Show graphic
plt.show()
```
